```python
import import_ipynb
from UNet import multi_unet_model #Uses softmax 

from tensorflow.keras.utils import normalize
import os
import glob
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
tr_list = os.listdir("data/")

train_images = []
train_masks = []
for file in tr_list:
    if file != '.DS_Store':
        folder = os.listdir("data/" + file + '/i')
        for i in folder:
            print(i)
            s = i[0:-4]
            m = s + '_m.png'
            print(m)
            img_hr = cv2.imread("data/" + file + '/i/' + i, 0)
            mask_hr = cv2.imread("data/" + file + '/m/' + m, 0)
            #print(mask_hr.shape)
            #img_hr = np.array(img_hr)
            img_hr = cv2.resize(img_hr, (80, 160), interpolation = cv2.INTER_AREA)
            #print(img_hr.shape)
            #img_hr = cv2.cvtColor(m, cv2.COLOR_BGR2RGB)
            train_images.append(img_hr)
            train_masks.append(mask_hr)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
n, h, w = train_masks.shape
train_masks_reshaped = train_masks.reshape(-1,1)
print(train_masks_reshaped.shape)
train_masks_reshaped_encoded = labelencoder.fit_transform(train_masks_reshaped)
print(train_masks_reshaped_encoded)
train_masks_encoded_original_shape = train_masks_reshaped_encoded.reshape(n, h, w)
print(train_masks_encoded_original_shape.shape)
```

```python
train_images = np.expand_dims(train_images, axis=3)
print(train_images.shape)
train_images = normalize(train_images, axis=1)

train_masks_input = np.expand_dims(train_masks_encoded_original_shape, axis=3)
print(train_masks_input.shape)
```

```python
np.unique(train_masks_encoded_original_shape)
```

```python
from sklearn.model_selection import train_test_split
X1, X_test, y1, y_test = train_test_split(train_images, train_masks_input, test_size = 0.10, random_state = 0)

X_train, X_do_not_use, y_train, y_do_not_use = train_test_split(X1, y1, test_size = 0.2, random_state = 0)

print(np.unique(y_train))
```

```python
y_train.shape
```

```python
n_classes = len(np.unique(y_train))
print(n_classes)

from tensorflow.keras.utils import to_categorical

train_masks_cat = to_categorical(y_train, num_classes=n_classes)
y_train_cat = train_masks_cat.reshape((y_train.shape[0], y_train.shape[1], y_train.shape[2], n_classes))

test_masks_cat = to_categorical(y_test, num_classes=n_classes)
y_test_cat = test_masks_cat.reshape((y_test.shape[0], y_test.shape[1], y_test.shape[2], n_classes))
```

```python
from sklearn.utils.class_weight import compute_class_weight
# class_weights = dict(zip(np.unique(train_masks_reshaped_encoded), compute_class_weight('balanced', classes = np.unique(train_masks_reshaped_encoded), 
#                 y = train_masks_reshaped_encoded)))

class_weights = compute_class_weight('balanced', classes = np.unique(train_masks_reshaped_encoded), y = train_masks_reshaped_encoded)
                     
# class_weights = compute_class_weight('balanced', classes = np.unique(train_masks_reshaped_encoded), y = train_masks_reshaped_encoded)
# class_weights = {i : class_weights[i] for i in range(8)}
print(class_weights)
```

```python
IMG_HEIGHT = X_train.shape[1]
IMG_WIDTH  = X_train.shape[2]
IMG_CHANNELS = X_train.shape[3]
print(n_classes)

print(X_train.shape)

def get_model():
    return multi_unet_model(n_classes=n_classes, IMG_HEIGHT=IMG_HEIGHT, IMG_WIDTH=IMG_WIDTH, IMG_CHANNELS=IMG_CHANNELS)
```

```python
import segmentation_models as sm

dice_loss = sm.losses.DiceLoss(class_weights=class_weights) 
focal_loss = sm.losses.CategoricalFocalLoss()
total_loss = dice_loss + (1 * focal_loss)
```

```python
import keras
model = get_model()
model.compile(optimizer='adam', loss=focal_loss, metrics=['accuracy'])
model.summary()
```

```python
print(len(class_weights))
history = model.fit(X_train, y_train_cat, 
                    batch_size = 16, 
                    verbose=1, 
                    epochs=25, 
                    validation_data=(X_test, y_test_cat), 
#                     class_weight=class_weights,
                    shuffle=False)
```

```python
model.save('UNet_Epochs_50.hdf5')
_, acc = model.evaluate(X_test, y_test_cat)
print("Accuracy is = ", (acc * 100.0), "%")
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

plt.plot(epochs, acc, 'y', label='Training Accuracy')
plt.plot(epochs, val_acc, 'r', label='Validation Accuracy')
plt.title('Training and validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
model.load_weights('UNet_Epochs_50.hdf5')  


y_pred=model.predict(X_test)
y_pred_argmax=np.argmax(y_pred, axis=3)
```

```python
from keras.metrics import MeanIoU
n_classes = 8
IOU_keras = MeanIoU(num_classes=n_classes)  
IOU_keras.update_state(y_test[:,:,:,0], y_pred_argmax)
print("Mean IoU =", IOU_keras.result().numpy())


#To calculate I0U for each class...
values = np.array(IOU_keras.get_weights()).reshape(n_classes, n_classes)
print(values)
class_IoU = []
for i in range (0,n_classes):
    IoU = values[i,i]/(values[i,i]+ values[i,0] + values[i,1] + values[i,2] + values[i,3] +  values[i,4] + values[i,5] + values[i,6]
                      +  values[i,7] + values[0,i] + values[1,i] + values[2,i] + values[3,i] +  values[4,i] + values[5,i] + values[6,i]
                      +  values[7,i])
    class_IoU.append(IoU)

print("IoU for class0 is: ", class_IoU[0])
print("IoU for class1 is: ", class_IoU[1])
print("IoU for class2 is: ", class_IoU[2])
print("IoU for class3 is: ", class_IoU[3])
print("IoU for class4 is: ", class_IoU[4])
print("IoU for class5 is: ", class_IoU[5])
print("IoU for class6 is: ", class_IoU[6])
print("IoU for class7 is: ", class_IoU[7])

plt.imshow(train_images[0, :,:,0], cmap='gray')
plt.imshow(train_masks[0], cmap='gray')
```
